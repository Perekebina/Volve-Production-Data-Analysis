```python
# # Import the daily oil, gas and water production from all producers wells in Volve
path1 = r'C:\Users\crosby\Documents\IFP School\Volve Database\Volve production data.xlsx'
df = pd.read_excel(path1, sheet_name='Daily Production Data')
```

```python
df.head()
```

```python
# Create a series for the GOR and WC
df['GOR'] = df['BORE_GAS_VOL']/df['BORE_OIL_VOL']
df['WC'] = df['BORE_WAT_VOL']*100/(df['BORE_OIL_VOL']+df['BORE_WAT_VOL'])
```

```python
df.columns
```

```python
well_list = df.WELL_BORE_CODE.unique()
```

```python
well_list
```

```python
df.set_index('WELL_BORE_CODE', inplace=True)
```

```python
# Create a plotting function to plot the oil rate and production uptime for any well in the well list
def plot_well(well):
    dff = df.loc[well]
    fig = go.Figure()
    fig.add_trace(
        go.Scatter(
            x=dff['DATEPRD'],
            y=dff['BORE_OIL_VOL'],
            mode='markers',
            marker = dict(
                color='green'
            ),
            name='Oil volume'
        )
    )
    fig.add_trace(
        go.Scatter(
            x=dff['DATEPRD'],
            y=dff['ON_STREAM_HRS'],
            mode='markers',
            name='On time hours',
            yaxis='y2'
            )
    )

    fig.update_layout(
        title='Daily Oil Volume and On-Time Hours for well: '+ well,
        xaxis=dict(
        title='Date'
        ),
        yaxis=dict(
            title="Oil Volume, sm3/d"
        ),
        yaxis2=dict(
            title="On Time, hrs",
            overlaying="y",
            side="right"
        )
    )
    fig.show()
    
well = widgets.Dropdown(options = well_list, description = 'Well Name:', disabled = False)

widgets.interactive(plot_well, well=well)
```

```python
def well_diag(well):
    dff = df.loc[well]

    fig = make_subplots(rows=3, cols=2, subplot_titles=("Oil Rate", "Gas Rate", "Water Rate", "Pressures", 'GOR', 'Water Cut'))

    fig.add_trace(
        go.Scatter(
            x=dff['DATEPRD'], 
            y=dff['BORE_OIL_VOL'],
            name='Oil', line=dict(color='green')),
            row=1, col=1
    )
    fig.add_trace(
        go.Scatter(
            x=dff['DATEPRD'], 
            y=dff['BORE_GAS_VOL'],
            name='Gas'),
            row=1, col=2
    )    
    
    fig.add_trace(
        go.Scatter(
            x=dff['DATEPRD'], 
            y=dff['BORE_WAT_VOL'],
            name='Water'),
            row=2, col=1
    )      
    fig.add_trace(
        go.Scatter(
            x=dff['DATEPRD'], 
            y=dff['AVG_DOWNHOLE_PRESSURE'],
            name='BHP'),
            row=2, col=2
    )
    fig.add_trace(
        go.Scatter(
            x=dff['DATEPRD'], 
            y=dff['AVG_WHP_P'],
            name='WHP'),
            row=2, col=2
    )    
    fig.add_trace(
        go.Scatter(
            x=dff['DATEPRD'], 
            y=dff['GOR'],
            name='GOR', line=dict(color='purple')),
            row=3, col=1
    )  
    fig.add_trace(
        go.Scatter(
            x=dff['DATEPRD'], 
            y=dff['WC'],
            name='Water Cut', line=dict(color='darkblue')),
            row=3, col=2
    )    

    fig.update_layout(height=600, width=800, title_text="Daily Production Diagnostic Plots")
    fig.show()
         
well = widgets.Dropdown(options = well_list, description = 'Well Name:', disabled = False)

widgets.interactive(well_diag, well=well)         
```

```python
# Create a series of the average oil rate and average uptime over the last six months for each well
well_list = monthly_prod['Wellbore name'].unique()
frames = []
dff = {}
for well in well_list:
    well_dff = monthly_prod[monthly_prod['Wellbore name'] == well]
    well_dff['Cum_Oil_Sm3'] = well_dff['Oil_Sm3'].cumsum()
    well_dff['Cum_Water_Sm3'] = well_dff['Water_Sm3'].cumsum()
    frames.append(well_dff)
dff = pd.concat(frames)
```
